```python
##Compute sum of digits of an integer (no string conversion).
```

```python
n = int(input("Enter number: "))
sum_digits = 0

while n > 0:
    sum_digits += n % 10   # last digit add
    n //= 10               # last digit remove

print("Sum of digits:", sum_digits)
```

```python
##Reverse a number (no string).

n = int(input("Enter number: "))
rev = 0

while n > 0:
    rev = rev * 10 + (n % 10)
    n //= 10

print("Reversed number:", rev)
```

```python
##Check palindrome number (no string).

n = int(input("Enter number: "))
temp = n
rev = 0

while n > 0:
    rev = rev * 10 + (n % 10)  # last digit add
    n //= 10                  # last digit remove

if temp == rev:
    print("Palindrome Number")
else:
    print("Not Palindrome Number")

```

```python
##Check Armstrong number (general n-digit).

n = int(input("Enter number: "))
temp = n
digits = 0

t = n
while t > 0:
    digits += 1
    t //= 10

sum_ = 0
while n > 0:
    d = n % 10
    sum_ += d ** digits
    n //= 10

if sum_ == temp:
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python
##Check Perfect number.

n = int(input("Enter number: "))
sum_ = 0

for i in range(1, n):
    if n % i == 0:
        sum_ += i

if sum_ == n:
    print("Perfect Number")
else:
    print("Not Perfect")
```

```python
##Check Strong number (sum of factorial of digits).

import math

n = int(input("Enter number: "))
temp = n
sum_ = 0

while n > 0:
    sum_ += math.factorial(n % 10)
    n //= 10

if sum_ == temp:
    print("Strong Number")
else:
    print("Not Strong")
```

```python
##Check Harshad/Niven number.

n = int(input("Enter number: "))
temp = n
sum_ = 0

while n > 0:
    sum_ += n % 10
    n //= 10

if temp % sum_ == 0:
    print("Harshad Number")
else:
    print("Not Harshad")
```

```python
##Check Automorphic number (ends with its square)

n = int(input("Enter number: "))
sq = n * n

temp = n
while temp > 0:
    if temp % 10 != sq % 10:
        print("Not Automorphic")
        break
    temp //= 10
    sq //= 10
else:
    print("Automorphic Number")
```

```python
##Check Neon number (sum digits of square equals number).

n = int(input("Enter number: "))
sq = n * n
sum_ = 0

while sq > 0:
    sum_ += sq % 10
    sq //= 10

if sum_ == n:
    print("Neon Number")
else:
    print("Not Neon")
```

```python
5##..Find GCD using Euclid algorithm.

a = int(input("Enter a: "))
b = int(input("Enter b: "))

while b != 0:
    a, b = b, a % b

print("GCD:", a)
```

```python
##.Find LCM using GCD

a = int(input("Enter a: "))
b = int(input("Enter b: "))
x, y = a, b

while b != 0:
    a, b = b, a % b

gcd = a
lcm = (x * y) // gcd

print("LCM:", lcm)
```

```python
##Power a^b mod m (Fast Exponentiation

a = int(input())
b = int(input())
m = int(input())

res = 1
a %= m

while b > 0:
    if b % 2 == 1:
        res = (res * a) % m
    a = (a * a) % m
    b //= 2

print(res)
```

```python
##Prime factors of a number

n = int(input("Enter number: "))
i = 2

while i * i <= n:
    while n % i == 0:
        print(i, end=" ")
        n //= i
    i += 1

if n > 1:
    print(n)
```

```python
##Check prime (âˆšn optimized)

n = int(input("Enter number: "))

if n <= 1:
    print("Not Prime")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
##Sieve of Eratosthenes

n = int(input("Enter limit: "))
prime = [True] * (n + 1)
prime[0] = prime[1] = False

for i in range(2, int(n**0.5) + 1):
    if prime[i]:
        for j in range(i*i, n+1, i):
            prime[j] = False

for i in range(2, n+1):
    if prime[i]:
        print(i, end=" ")
```

```python
##Twin primes in range

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input("Enter limit: "))
for i in range(3, n):
    if is_prime(i) and is_prime(i+2):
        print(i, i+2)
```

```python
##Factorial using recursion

def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n-1)

print(fact(int(input())))
```

```python
## Compute nCr (combinations) using factorial + simplification.

def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

n = int(input("Enter n: "))
r = int(input("Enter r: "))

if r > n:
    print("nCr not defined")
else:
    r = min(r, n - r)   # simplification
    result = factorial(n) // (factorial(r) * factorial(n - r))
    print("nCr =", result)
```

```python
## sin(x) using Taylor series

import math

x = float(input("x in radians: "))
n = int(input("terms: "))
res = 0
sign = 1

for i in range(n):
    res += sign * (x ** (2*i+1)) / math.factorial(2*i+1)
    sign *= -1

print(res)
```

```python
## e^x using Taylor series

import math

x = float(input())
n = int(input())
res = 0

for i in range(n):
    res += (x ** i) / math.factorial(i)

print(res)
```
